```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
import numpy as np
import pandas as pd
```

```python
#import iris dataset
from sklearn.datasets import load_iris
import pandas as pd

# Load the iris dataset
iris = load_iris()
iris
```

```python
#create data frame
df=pd.DataFrame(data=iris.data,columns=iris.feature_names)
```

```python
#add new column "Species"
df['Species']=pd.Categorical.from_codes(iris.target,iris.target_names)
```

```python
#print a header msg and ensure all rows are in string and print it
print("iris Dataset(all rows):\n")
print(df.to_string(index=False))
```

```python
#save to excel file
df.to_excel('iris.xlsx',index=False)
```

```python
#Q2 Visualize the first two features;sepal length vs sepal width
plt.figure(figsize=(6,4))
sns.scatterplot(data=df,x='sepal length (cm)',y='sepal width (cm)',hue="Species",palette='Set1')
plt.title("sepal length vs sepal width")
plt.xlabel("sepal length(cm)")
plt.ylabel("sepal width(cm)")
plt.legend()
plt.grid(True)
plt.tight_layout() #adjust spacing
plt.show()
```

```python
#Perform kmeans
X=iris.data[:,:2] #select first two features
#k=3
kmeans=KMeans(n_clusters=3)
kmeans.fit(X)
#add cluster labels to data frame
df['Sepal_Cluster']=kmeans.labels_
```

```python
#visualise clusters
plt.scatter(X[:,0],X[:,1],c=kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],c='red',marker='*',s=200)
plt.show()
```

```python
#next two features
Y=iris.data[:,2:4]
#apply kmeans k=3
kmeans1=KMeans(n_clusters=3,random_state=0)
kmeans1.fit(Y)
#add cluster labels
df['Petal_cluster']=kmeans1.labels_
```

```python
#visualize clusters
plt.scatter(Y[:,0],Y[:,1],c=kmeans1.labels_)
plt.scatter(kmeans1.cluster_centers_[:,0],kmeans1.cluster_centers_[:,1],c='red',marker='*',s=200)
plt.show()
```

```python
#print final cluster
print("\nIris Dataset with assigned clusters:\n")
print(df.to_string(index=False))
```

```python
#save to excel
df.to_excel("iris clusters.xlsx",index=False)
print("\nDataset with clusters saved as 'iris_clusters.xlsx'")
```
