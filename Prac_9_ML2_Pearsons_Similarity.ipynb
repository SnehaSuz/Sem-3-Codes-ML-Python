```python
import numpy as np
import pandas as pdimport warnings
warnings.filterwarnings('ignore')
```

```python
#User item matrix
ratings_dict={
    "I1":[5,3,4,3],
    "I2":[4,1,3,3],
    "I3":[1,2,4,1],
    "I4":[4,3,3,5],
    "I5":[np.nan,3,5,4]
}
users=["Alice","U1","U2","U3"]
ratings=pd.DataFrame(ratings_dict,index=users)
print("===User_Item Ratings Matrix===")
print(ratings)

```

```python
#Compute item-item similarity
item_similarity_df=ratings.corr(method="pearson")
print("\n===Item-Item Similarity Matrix===")
print(item_similarity_df.round(2))
```

```python
#Predict missing rating
target_user="Alice"
target_item="I5"

rated_items=ratings.loc[target_user].dropna()

numerator,denominator=0,0
for item,rating in rated_items.items():
  sim=item_similarity_df.loc[target_item,item]
  if not np.isnan(sim):
    numerator+=sim*rating
    denominator+=abs(sim)

predicted_rating=numerator/denominator if denominator !=0 else 0
print(f"\nPredicted rating of {target_user} for {target_item}:{predicted_rating:.2f}")
```

```python
#Generate top n rec
def recommend_items_item_based(user,n=2):
  user_ratings=ratings.loc[user]
  predictions={}

  for item in ratings.columns:
    if pd.isna(user_ratings[item]):
      num,den=0,0
      for other_item,rating in user_ratings.dropna().items():
        sim=item_similarity_df.loc[item,other_item]
        if not np.isnan(sim):
          num+=sim*rating
          den+=abs(sim)
      predictions[item]=num/den if den !=0 else 0

      return sorted(predictions.items(),key=lambda x:x[1],reverse=True)[:n]


print(f"\n===Top Reccomendation for {target_user}===")
print(recommend_items_item_based(target_user,n=2))
```
