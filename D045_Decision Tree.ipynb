```python
#define categorical and numerical columns
#our target variable is present in dataset i.e Quality
categories=df1.select_dtypes(include='object')
numerical=df1.select_dtypes(exclude='object').drop(columns=['Survived'])
```

```python
#we find out unique values in column so that we can move on to encoding
categories.nunique()
```

```python
le=LabelEncoder()
df1['Survived']=le.fit_transform(df1['Survived'])
```

```python
ohe=OneHotEncoder(sparse_output=False).set_output(transform='pandas')
encoded=ohe.fit_transform(categories)
encoded
```

```python
final_df=pd.concat((df1,encoded),axis=1).drop(columns=categories)
final_df
```

```python
#define x and y
x=final_df.drop(columns=['Survived'])
y=final_df['Survived']
```

```python
#split the dataset
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#Deploy the model
model=DecisionTreeClassifier()
model.fit(X_train,y_train)
```

```python
#predicting survival
y_pred=model.predict(X_test)
#evaluation
report=classification_report(y_test,y_pred)
print("\nClassification Report:\n",report)
```
