```python
import pandas as pd
```

```python
data=pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/refs/heads/master/titanic.csv')
data
```

```python
data.info()
```

Dataset has 7 numerical columns, 5 categorical columns

```python
data.describe()
```

```python
#check unique values in CABIN
data['Cabin'].unique()
data
```

```python
#unique values in parch
data['Parch'].unique()
#
data['SibSp'].unique()
```

```python
data.shape
```

```python
#removing column: ticket, embarked,name, PassengerId	(columns not needed)
data.drop(['Ticket','Embarked','Name','PassengerId','Cabin'],axis=1,inplace=True)
data
```

```python
#convert sex into float
data['Sex']=data['Sex'].map({'male':0,'female':1})
data
```

```python
#REPLACING NULL AGES BY MEAN AGE VALUE
data['Age'].fillna(data['Age'].mean(),inplace=True)
data
```

```python
#separate input x and out y
x=data.iloc[:,1:]
y=data.iloc[:,0]
data

```

```python
#spliting train and test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
x_train
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
#predict data and store
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
accuracy_score(y_test,y_pred)
```

```python
#feature importance
coefficients=pd.DataFrame({'Feature':x.columns,'Coefficients':model.coef_[0]})
coefficients.sort_values(by='Coefficients',ascending=False,inplace=True)
coefficients
```

```python
#plot barplot fot feature importance
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.barh(coefficients['Feature'],coefficients['Coefficients'])
plt.xlabel('Coefficients')
plt.ylabel('Features')
plt.title('Feature Importance')
plt.show()
```

```python
#import adaboost
from sklearn.ensemble import AdaBoostClassifier
abs=AdaBoostClassifier()
abs.fit(x_train,y_train)
```

```python
#train data fot this model
y_pred=abs.predict(x_train)
accuracy_score(y_train,y_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf=AdaBoostClassifier(random_state=100,estimator=RandomForestClassifier(random_state=10),n_estimators=100,learning_rate=0.005)
clf.fit(x_train,y_train)
```

```python
#predict accuracy
y_pred=clf.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=AdaBoostClassifier(random_state=100,estimator=DecisionTreeClassifier(random_state=43),n_estimators=100,learning_rate=0.05)

dt.fit(x_train,y_train)
y_pred=dt.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
#XGboost
from xgboost import XGBClassifier
xgb=XGBClassifier()
xgb.fit(x_train,y_train)
y_pred=xgb.predict(x_test)
accuracy_score(y_test,y_pred)
```
