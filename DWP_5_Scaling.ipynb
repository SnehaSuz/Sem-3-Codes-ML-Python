Normalization: Its a scalling technique in which values are shifted and
rescaled ranging between 0 and 1. Also known as Min-Max Scaler.
Formula is, x'=(x-xmin)/(xmax-xmin).


Standardization is another scaling technique where values centered around mean with a unit standard deviation(1sigma,2sigma,3sigma). Follows empirical rule.

Formula, x'=(x-miu)/sd

If our data is not normally distributed , we use Normalization. If data is normally distribution standardization is used.

```python
import pandas as pd
df=pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/refs/heads/master/BostonHousing.csv')
df
```

```python
#min max and standard
from sklearn.preprocessing import MinMaxScaler,StandardScaler
scaler=MinMaxScaler()#make an object of class
normalize_data=scaler.fit_transform(df)
normalize_data
```

```python
#name the columns
new_df=pd.DataFrame(normalize_data,columns=df.columns)
new_df
```

```python
new_df.describe()
```

Our data is between 0 to 1 now. Before pca apply scaling technique.

```python
#standard scaler
scaler=StandardScaler()
standard_data=scaler.fit_transform(df)
standard_data
```

```python
#name columns
new_df1=pd.DataFrame(standard_data,columns=df.columns)
new_df1
```

```python
new_df1.describe()
```

MEAN NORMALIZATION
x'=x-mean/xmax-xmin

```python
#apply this formula on the data set
mean_norm=(df-df.mean())/(df.max()-df.min())
mean_norm
```

MAXIMUM ABSOLUTE SCALER: used when data has 0 centered distribution i.e most of the data lies near mean

formula is x'=x/max|x|

```python
#apply this amx absolute scaler
from sklearn.preprocessing import MaxAbsScaler
scaler=MaxAbsScaler()
max_abs_data=scaler.fit_transform(df)
max_abs_data
#make colums
new_df2=pd.DataFrame(max_abs_data,columns=df.columns)
new_df2

#x'=x/(max|x|)
```

To deal witth outliers we have MEDIAN & QUANTILE SCALING

Formula= x'=x-med(x)/IQR(x)

```python
#import and apply
from sklearn.preprocessing import RobustScaler
scaler=RobustScaler()
robust_data=scaler.fit_transform(df)
robust_data
#make columns
new_df3=pd.DataFrame(robust_data,columns=df.columns)
new_df3
```

```python
new_df3.describe()
```

REGULARIZATION : L2 Regularization is also called as Ridge Regularization.It prevents the over fitting problem by adding penalty to the loss function. This penalty discourages large coefficients in the model i.e variables(columns) with maximum magnitudes


VECTOR UNIT LENGTH SCALING using L2 NORM

```python
#import for l2 normalize
from sklearn.preprocessing import Normalizer
normalizer=Normalizer(norm='l2')
l2_data=normalizer.fit_transform(df)
l2_data
#make columns
new_df4=pd.DataFrame(l2_data,columns=df.columns)
new_df4
```
