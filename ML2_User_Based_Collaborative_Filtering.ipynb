USER BASED COLABORATIVE FILTERING

```python
import numpy as np
import pandas as pd

np.random.seed(42)
users = [f'User_{i}' for i in range(1, 101)]
movies = [f'Movie_{i}' for i in range(1, 51)]

ratings_data = []
for user in users:
    for movie in movies:
        if np.random.random() > 0.7:
            rating = np.random.randint(1, 6)
            ratings_data.append({'user': user, 'movie': movie, 'rating': rating})

df = pd.DataFrame(ratings_data)
user_item_matrix = df.pivot(index='user', columns='movie', values='rating')

def pearson_similarity(matrix):
    matrix_centered = matrix.sub(matrix.mean(axis=1), axis=0)
    similarity = matrix_centered.T.corr(min_periods=2)
    return similarity.fillna(0)

user_similarity_pearson = pearson_similarity(user_item_matrix)

print("User-User Pearson Similarity (first 5 users):\n")
print(user_similarity_pearson.head())

def get_top_k_similar_users(user, k=5):
    if user not in user_similarity_pearson.index:
        return []
    sims = user_similarity_pearson.loc[user].drop(user)
    return sims.nlargest(k)

print("\nTop 5 similar users to User_1 (Pearson):")
print(get_top_k_similar_users("User_1", k=5))
```
